5. Aprovação de aluno
Peça a nota do aluno. Se a nota for maior ou igual a 7,
mostre "Aluno aprovado". Caso contrário, mostre "Aluno
reprovado". (Use if e else).


```python
nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    print("Aluno aprovado")
else:
    print("Aluno reprovado")
```
