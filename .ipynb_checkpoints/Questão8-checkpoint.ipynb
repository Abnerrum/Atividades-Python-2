8. Nota do aluno
Peça a nota do aluno. Mostre:
▶ Aprovado (nota ≥ 7)
▶ Recuperação (nota entre 5 e 6)
▶ Reprovado (nota < 5)
(Use if, elif e else).


```python
nota = float(input("Digite a nota do aluno: "))

if nota >= 7:
    print("Aprovado")

elif nota >= 5:
    print("Recuperação")

else:
    print("Reprovado")
```
