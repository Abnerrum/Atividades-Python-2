7. Classificação de idade
Peça a idade da pessoa. Mostre:
▶ Criança (0–12)
▶ Adolescente (13–17)
▶ Adulto (18 ou mais)
(Use if, elif e else).

```python
idade = int(input("Digite a idade: "))

if idade <= 12:
    print("Criança")

elif idade <= 17:
    print("Adolescente")

else:
    print("Adulto")
```
