1. Verificar número positivo
Peça ao usuário um número e verifique: Se o número for
maior que 0, mostre "Número positivo". (Use apenas
if).

```python
numero = int(input("Digite um número: "))

if numero > 0:
    print("Número positivo")

elif numero < 0:
    print("Número negativo")

else:
    print("Número é zero")
```
